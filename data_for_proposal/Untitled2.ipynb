```python
import pandas as pd
```

source: The world bank: Agriculture & Rural Development

```python
full_data = pd.read_csv('API_1_DS2_en_csv_v2_5873480.csv')
full_data.head()
```

```python
data = full_data[full_data['Country Name'] == 'Cambodia']
```

```python
data.head()
```

```python
data['Indicator Name'].unique()
```

```python
data.drop(['Country Name', 'Country Code', 'Indicator Code'], axis=1, inplace=True)
data.set_index('Indicator Name', inplace=True)
data.head()
```

```python
data = data.T
data.head()
```

```python
data.index
```

```python
data.loc['Unnamed: 67', :]
```

```python
data.drop('Unnamed: 67', axis=0, inplace=True)
```

```python
data.index = pd.to_datetime(data.index, format='%Y')
```

```python
data.head()
```

```python
filter_date = pd.to_datetime('2010', format='%Y')
data_gt_2010 = data[data.index >= filter_date]
data_gt_2010
```

```python
pd.set_option('display.max_columns', None)
data_gt_2010[data_gt_2010.index == '2022-01-01']
```

```python
data_gt_2010.drop('2022-01-01', axis=0, inplace=True)
```

```python
data_gt_2010
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 12))

missing_value = data_gt_2010.isnull().sum() / len(data_gt_2010)
missing_value.sort_values(inplace = True)

sns.barplot(x=missing_value, y=missing_value.index)
```

```python
data_gt_2010['Employment in agriculture (% of total employment) (modeled ILO estimate)']
```

```python
filter_columns = missing_value[missing_value > 0].index
```

```python
# Drop some columns
data_gt_2010.drop(filter_columns, axis=1, inplace=True)
data_gt_2010
```

```python
sns.lineplot(data_gt_2010['Fertilizer consumption (kilograms per hectare of arable land)'])
```
